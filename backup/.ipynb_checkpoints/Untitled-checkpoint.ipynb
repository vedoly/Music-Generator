```python
def get_feedforward():
    input1 = tf.keras.layers.Input(shape=(10,3))
    x = tf.keras.layers.Dense(100, activation='relu')(input1)
    x = tf.keras.layers.Dense(100, activation='relu')(x)
    x = tf.keras.layers.Dense(100, activation='relu')(x)
    out = tf.keras.layers.Dense(3, activation='relu')(x)
    
    model = tf.keras.models.Model(inputs=input1, outputs=out)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(lr=1e-4),
        loss='mean_squared_error',
        metrics=['acc']
    )
    return model
```

```python
model = get_feedforward()
model.summary()
```

```python
model.fit(x_train, y_train, epochs=3, batch_size=1, verbose=1)
```
